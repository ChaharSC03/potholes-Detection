```python
import os
from glob import glob
import pandas as pd
from functools import reduce
```

```python
from xml.etree import ElementTree as et
```

```python
# load all xml files
xml_lists=glob('./data/annotations/*.xml')
```

```python
xml_lists
```

```python
xmlfiles=list(map(lambda x: x.replace('\\','/'),xml_lists))
```

```python
xmlfiles
```

```python
import xml.etree.ElementTree as ET

# Parse the XML document (you can replace xml_data with your actual XML data)
# root = ET.parse(xmlfiles)  # Make sure xml_data contains your XML
def extract_txt(filename):
    trees=et.parse(filename)
    root=trees.getroot()
    image_name = root.find('filename').text
    width = root.find('size').find('width').text
    height = root.find('size').find('height').text

    objects = root.findall('object')
    parser = []

    for obj in objects:
        name = obj.find('name').text
        bndbox = obj.find('bndbox')  
        xmin = bndbox.find('xmin').text
        xmax = bndbox.find('xmax').text
        ymin = bndbox.find('ymin').text
        ymax = bndbox.find('ymax').text

        parser.append([image_name, width, height, name, xmin, xmax, ymin, ymax])

    return parser
```

```python
parserAll=list(map(extract_txt,xmlfiles))
```

```python
parserAll
```

```python
data = reduce(lambda x,y :x+y, parserAll)
```

```python
data
```

```python
df=pd.DataFrame(data,columns=['filename','width', 'height', 'name','xmin','xmax','ymin','ymax'])
```

```python
df.info()
```

```python
# type conversion

cols=['width','height','xmin','xmax','ymin','ymax']

df[cols]=df[cols].astype(int)
df.info()
```

```python
# centerX and centerY

df['centerX']=((df['xmax']+df['xmin'])/2)/df['width']
df['centerY']=((df['ymax']+df['ymin'])/2)/df['height']
```

```python
# w
df['w']=(df['xmax'] - df['xmin'])/df['width']
df['h']=(df['ymax'] - df['ymin'])/df['height']
```

```python
iamges= df['filename'].unique()

len(iamges)
```

```python
imgDF=pd.DataFrame(iamges, columns=['filename'])
img_train=tuple(imgDF.sample(frac=0.8)['filename'])
```

```python
len(img_train)
```

```python
img_test= (imgDF.query(f'filename not in {img_train}')['filename'])
```

```python
train_df = df[df['filename'].isin(img_train)]
test_df = df[df['filename'].isin(img_test)]
```

```python
def label_encode(x):
    labels={'pothole':0}

    return labels[x]
```

```python
train_df['id']=train_df['name'].apply(label_encode)
train_df
```

```python
test_df['id']=test_df['name'].apply(label_encode)
test_df.info
```

```python
from shutil import move
```

```python
train_folder= 'data/train'
test_folder='data/test'

os.mkdir(train_folder)
os.mkdir(test_folder)
```

```python
col=['filename','id','centerX',	'centerY','w','h']

groupby_obj_train=train_df[col].groupby('filename')
groupby_obj_test=test_df[col].groupby('filename')
```

```python
groupby_obj_train.get_group('potholes1.png')
```

```python
def save_data(filename, folder, group_obj):
    src=os.path.join('data/images/',filename)
    dst=os.path.join(folder, filename)
    move(src,dst)

    txt_filename=os.path.join(folder,os.path.splitext(filename)[0]+'.txt')

    group_obj.get_group(filename).set_index('filename').to_csv(txt_filename,sep=' ',index=False, header=False)
```

```python
groupby_obj_train.groups
```

```python
files_series=pd.Series(groupby_obj_train.groups.keys())
```

```python
files_series.apply(save_data,args=(train_folder, groupby_obj_train))
```

```python
files_series_test=pd.Series(groupby_obj_test.groups.keys())

files_series_test.apply(save_data,args=(test_folder, groupby_obj_test))
```
